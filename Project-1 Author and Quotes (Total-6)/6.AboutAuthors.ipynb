```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from tqdm import tqdm
```

### For this to work first we need to extract all the authors and their about page link from the quotes
### Then we need to select the distinct ones
### then we need to save them in multiple files in a loop

```python
link="https://quotes.toscrape.com/"
```

```python
Author=list()                        # here the author name and the about page link will be extracted and stored
```

```python
def extractAuthor(link):
    result=requests.get(link)
    soup=BeautifulSoup(result.text,'html.parser')
    for detail in soup.find_all('div',class_='quote'):
        author=detail.find('small',class_='author').text
        about= detail.find('a').get('href')
        Author.append([author,"https://quotes.toscrape.com/"+about])
    nextlink= soup.find('li',class_='next')
    if nextlink is not None:
        link=nextlink.find('a').get('href')
        link= "https://quotes.toscrape.com/"+link
        extractAuthor(link)

    
```

```python
extractAuthor(link)
```

```python
AD= pd.DataFrame(Author,columns=['Author','link'],index=None)
```

```python
AD.drop_duplicates(inplace=True)
```

```python
ADD= np.array(AD)
```

```python
ADD
```

```python
a=1
for detail in tqdm(ADD):
    link= str(detail[1])
    result= requests.get(link)
   
    soup=BeautifulSoup(result.text,'html.parser')
    name= "Authors/" + str(a)+". "+str(detail[0])+".txt"
    file= open(name,'w',encoding='utf-8')
    file.write(soup.find('div',class_='author-details').text)
    file.close()
    a+=1
```
