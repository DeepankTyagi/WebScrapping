```python
import requests
import numpy as np
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
link = "https://quotes.toscrape.com/"
```

```python
AllDetails= list()
```

```python
def extract(link):
    result= requests.get(link)
    soup= BeautifulSoup(result.text,'html.parser')
    Details= soup.find_all('div',class_='quote')
    for detail in Details:
        quote= detail.find('span',class_='text').text[1:-1]
        author= detail.find('small',class_='author').text
        about= "https://quotes.toscrape.com/"+detail.find('a').get('href')
        tags= list()
        for tag in detail.find_all('a',class_='tag'):
            tags.append(tag.text)
        ",".join(tags)
        AllDetails.append([quote,','.join(tags),author,about])
    n = soup.find('li',class_='next')
    if n is not None:
        next= n.find('a').get('href')
        link="https://quotes.toscrape.com/"+next                           
        print(link)                                                   # To check where it is working properly or not
        extract(link)
    else:
        return 0
    
    
```
